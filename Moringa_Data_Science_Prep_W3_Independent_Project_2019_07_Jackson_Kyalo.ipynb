# Data Analysis and Exploration

```python
#Which ones were the most used city for the three days?
#The cities with the highest value counts have the most used
#displaying the 5 most used cities
merged['City'].value_counts().head()
```

```python
#Which cities were the most used during business and home hours?
#Business Hours in Ivory Coast are between 8am and 4 pm
#To answer this we shall get the most city with the highest cell usage during business hours and home hours.  
merged[['Time']].groupby([merged['Cell_ID'], merged['Time']]).count()
```

```python
#Most used city for the three days?
#To get this we will get the sum of calls grouped by city 
#and the one with the max number of calls is the most used city
grouped=merged['Cell_ID'].groupby(merged['City']).count()
grouped[grouped==grouped.max()]
```

```python
#Understanding the business cell 
merged.groupby('Cell_ID').count().nlargest(10,'Product')
```
